### Question 5 (5pts)
Write a function that fits these regression estimators to the training data using a percentage-split approach. You should use an iteration statement to apply these regression estimators. R-squared values should be displayed for both training and test sets.

```python
from sklearn.metrics import r2_score

def regressors_percentage_split():
    """Fits the above regression estimators to the training data using a percentage-split approach"""
    
    # iterate through the dictionary keys
    for it in estimators:
        # print key/regression estimator name
        print("\n",it,":",sep="")
        
        # fit the model for this estimator
        model = estimators[it].fit(X=X_train, y=y_train)


        # print the r2 results
        print("\tR-squared value for training set: ", r2_score(y_train, model.predict(X_train)))
        print("\tR-squared value for testing set: ", r2_score(y_test, model.predict(X_test)))
```

### Question 6 (6pts)

For this question, you will normalize the X_train and X_test variables and perform the same multiple model fitting/evaluation that you did in the previous question. The output of this question will be similar to that of the previous question, with only one major distinction: here the normalized variables will be used to train and test the algorithms. For normalization, apply **MinMaxScaler()**.

Write a function that first applies the MinMaxScaler to the X_train and X_test global variables (as defined above) and then applies the same regression estimators. This function should also return the scaled variables, X_train_scaled and X_test_scaled respectively. You will need these variables in the next question.

```python
from sklearn.preprocessing import MinMaxScaler

def preprocessed_regression():
    """Normalize the X_train and X_test variables above and then fits the above regression 
    estimators to that training data using a percentage-split approach"""

    # create and fit the scaler
    scaler = MinMaxScaler()
    scaler.fit(X_train) 
    
    # transforming the X data
    X_train_scaled = scaler.transform(X_train) 
    X_test_scaled = scaler.transform(X_test) 
    
    # iterate through the dictionary keys
    for it in estimators:
        # print key/regression estimator name
        print("\n",it,":",sep="")
        
        # fit the model for this estimator
        model = estimators[it].fit(X=X_train_scaled, y=y_train)

        # print the r2 results
        print("\tR-squared value for training set: ", r2_score(y_train, model.predict(X_train_scaled)))
        print("\tR-squared value for testing set: ", r2_score(y_test, model.predict(X_test_scaled)))
    
    
    return X_train_scaled, X_test_scaled
```

```python
X_train_scaled, X_test_scaled = preprocessed_regression()
```

### Question 7 (6pts)
Looking at the previous output, it can be said that the kNN algorithm seems to works best for this dataset. In this question, you will attempt to improve the prediction performance by automatically selecting the most important features (k=3). 

Use recursive feature elimination (RFE) to select the **three** most important features. For your RFE, use a decision tree regressor to recursively eliminate the features. Your selection must be based on the X_train_scaled and X_test_scaled variables defined in the previous question. 

The idea is to use the scaled values that provided better results than raw values and to see if we can further improve the prediction performance by only using the most important features. Your function should produce the sample output given below and should also return the selected features, X_train_selected and X_test_selected.

```python
from sklearn.feature_selection import RFE
from sklearn.tree import DecisionTreeRegressor

def RFE_feature_selection():
    """Uses recursive feature elimination (RFE) to select the three most important features of the above
    data and fits the data to the kNN regression estimator."""
    
    # create the RFE and fit it to the scaled data
    select = RFE(DecisionTreeRegressor(random_state = 3000),n_features_to_select = 10)
    select.fit(X_train_scaled, y_train)

    # transforming the X data
    X_train_selected = select.transform(X_train_scaled)
    X_test_selected = select.transform(X_test_scaled)

    # fit the kNN model with the transformed data
    model = KNeighborsRegressor().fit(X=X_train_selected, y=y_train)
    
    # print the selected 3 features
    print("Selected features after RFE:")
    for i in range(len(select.get_support())):
        if select.get_support()[i]:
            print("\t",features.columns[i],sep="")

    # print the kNN performance data
    print("\nkNN Regression performance with selected features:")
    print("\tR-squared value for training set: ", r2_score(y_train, model.predict(X_train_selected)))
    print("\tR-squared value for testing set: ", r2_score(y_test, model.predict(X_test_selected)))
    
    return X_train_selected, X_test_selected
```

```python
X_train_selected, X_test_selected = RFE_feature_selection()
```

### Question 8 (5pts)

In this last question, you will perform a Grid Search to tune two key parameters of the kNN regression algorithm at the same time and determine which combination yields the best performance. Specifically, you will change the number of neighbors, **n_neighbors**, and the distance function used to calculate the distance between different data points, **metric**. The possible values for each hyperparameter are as follows:
    * For n_neighbors, try 1, 5, and 10.
    * For metric, try ""euclidean", "manhattan", and "minkowski".

You can learn more about these three distance functions at https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.DistanceMetric.html.


Write a function that applies a Grid Search with kNN regression estimator and evaluates the performance of these parameters using 5-fold evaluation. Your function should produce the provided output. Your function should use the X_train_selected and X_test_selected variables. 

```python
from sklearn.model_selection import GridSearchCV

def grid_search_kNN():
    """Performs a grid search to fine tune the n_neighbors and metric of the above kNN model."""
    
    # perform and fit the above grid search
    grid_search = GridSearchCV(KNeighborsRegressor(),{"n_neighbors":[1,5,10],"metric":["euclidean", "manhattan", "minkowski"]},cv=5)
    grid_search.fit(X=X_train_selected, y=y_train)
    
    # outputs results
    print("Best parameters: ", grid_search.best_params_)
    print("Training set score with best parameters: ", grid_search.score(X_train_selected, y_train))
    print("Test set score with best parameters: ", grid_search.score(X_test_selected, y_test))
```
